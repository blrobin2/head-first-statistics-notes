```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Pie Chart

```python
units_sold_by_genre = pd.DataFrame(data={'Units Sold': [27500, 11500, 6000, 3500, 1500], 'Genre': ['Sports', 'Strategy', 'Action', 'Schooter', 'Other'] })
```

```python
plt.pie(
    units_sold_by_genre['Units Sold'],
    labels=units_sold_by_genre['Genre'] + "\n" + units_sold_by_genre['Units Sold'].astype('str'),
    labeldistance=1.2,
    autopct='%.0f%%'
);
```

# Bar Chart

```python
df = pd.DataFrame(data = {'Continent': ['North America', 'South America', 'Europe', 'Asia', 'Oceania', 'Africa', 'Antartica'], 'Sales (Units)': [1500, 500, 1500, 2000, 1000, 500, 1]})
```

```python
df.index = df['Continent']
```

```python
df['Sales (Units)'].plot(kind='barh', legend=True).invert_yaxis();
```

* __categorical data__: data that is split into categories that describe qualities or characteristics (also called __qualitative__ date)
    * Example: game genre, breed of dog, types of dessert
    * The data values can't be interpreted as numbers
* __numerical data__: data that deals with _numbers_, such as measurements or counts (also called __quantitative__ data)
    * Examples: weight, length, time

# Histogram

Similar to bar charts, but:
1. The area of each bar is proportional to th frequency
2. There are no gaps between the bars on the chart

```python
manic_mango_scores = np.concatenate(
    (
        np.random.randint(low=0, high=199, size=5),
        np.random.randint(low=200, high=399, size=29),
        np.random.randint(low=400, high=599, size=56),
        np.random.randint(low=600, high=799, size=17),
        np.random.randint(low=800, high=999, size=3)
    )
)
```

```python
scores = pd.DataFrame(manic_mango_scores)
scores.plot(kind='hist', bins=[0,200,400,600,800,1000], legend=False)
```

### Summary So Far

* The __frequency__ is a statistical way of saying how many items there are in a category
* __Pie charts__ are good for showing basic proportions
* __Bar charts__ give you more flexibility and precision
* __Numerical data__ deals with numbers and quantities; __categorical data__ deals with words and qualities
* __Horizontal bar charts are used for categorical data__, particularly when the category names are long
* __Vertical bar charts are used for numerical data__, or categorical data if the category names are short
* You can show __multiple sets of data on a bar chart__, and you have a choice of how to do this
    * You can compare frequencies by showing related bars side-by-side on a _split category_ bar chart
    * You can show proportions and total frequencies by stacking the bars on top of each other on a _segmented_ bar chart
* __Bar chart scales__ can show either percentages or frequencies
* Each chart comes in a number of different varieties

## Frequency Density

```python
hours_played = np.concatenate((
    np.random.uniform(low=0.0, high=0.99, size=4300),
    np.random.uniform(low=1.0, high=2.99, size=6900),
    np.random.uniform(low=3.0, high=4.99, size=4900),
    np.random.uniform(low=5.0, high=9.99, size=2000),
    np.random.uniform(low=10.0, high=24.0, size=2100)
))
hour_bins = [0,1,3,5,10,24]
```

```python
hours_played_df = pd.DataFrame(hours_played, columns=['Hours'])
hours_played_df.plot(kind='hist', bins=hour_bins, edgecolor='black')
```

Because of the irregular shaped bins, 10-24 for example looks more prominent than it actually is...

```python
hours_played_df.plot(kind='hist', bins=hour_bins, edgecolor='black', density=True);
```

* __Frequency density__ relates to how concentrated the frequencies are for grouped data. It is calculated using:
    * $Frequency\,Density = \frac{Frequency}{Group\,width}$
* A __histogram__ is a chart that specializes in grouped data. It looks like a bar chart, but the hight of each bar equates to the frequency density rather than frequency
* When drawing histograms, the width of each bar is proportional to the width of its group. The bars are shown on a continuous numeric scale
* In a histogram, the frequency of a group is given by the area of its bar
* A histogram has no gaps between its bars

## Cumulative Frequency

The total frequency up to certain value. It's basically a running total of the frequencies

```python
counts, bins = np.histogram(hours_played, bins=hour_bins)
print(f"Count at 4 hours is ~{counts[0] + counts[1] + (counts[2] / 2)}")
plt.plot(bins[1:], counts.cumsum());
```

* __Cumulative frequency__ is the total frequency up to a particular value. It's a running total of the frequencies
* Use a cumulative frequency graph to plot the upper limit of each group of data against cumulative frequency
* Use a line chart if you want to show trends, for example over time
* You can show more than one set of data on a line chart
    * Use one line for each set of data, and make sure it's clear which line is which
* You can use line charts to make basic predictions as it's easy to see the shape of the trend. Just extend the trend line, trying to keep the basic shape
* __Don't use line charts to show categorical data__ unless you're showing trends for each category, for example over time
    * If you do this, draw one line per category
